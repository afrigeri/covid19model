```python
import requests
import pandas as pd
import seaborn as sns
import io
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
DATA="https://covid.ourworldindata.org/data/full_data.csv"
```

```python
s=requests.get(DATA).content
ww_data=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
ch_it_data = ww_data[(ww_data.location=='China') | (ww_data.location=='Italy')]
```

```python
plt.figure(figsize=(16,10))
chart = sns.lineplot(data=ch_it_data, x='date', y='total_cases', hue='location')
plt.xticks(rotation=65, horizontalalignment='right')
ax = plt.axes()        
ax.xaxis.grid() # vertical lines
```
